```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
a,b,c=int(input('  x = ?')), int(input('  y = ?')), int(input('  y = /?'))
t=np.arange(0, 100, 0.01)
x = np.sin(math.pi*t/a)
y = np.sin((math.pi*t+math.pi/c)/b)
plt.plot(x, y)
plt.xlabel(' X')
plt.ylabel(' Y')
plt.grid(True, linestyle='dotted')
plt.title(' ')
plt.show()
```

```python
plt.pie([0.01, 0.5, 0.3],autopct='%1.f%%' , labels = [',   ',',       ','      '])
plt.title('     ')
plt.show()
```

```python
a=np.arange(0,2,0.01)
b=np.sin(a*math.pi)
plt.grid(True, linestyle='dotted')
plt.hist(b, 100)
plt.title('  ')
plt.show()
```
