Задача A

```python
a=input()
def func(a):
    stc=[]
    otk=['(','[','{']
    zak=[')',']','}']
    for i in range(len(a)):
        if a[i] in otk:
            stc.append(a[i])
            if i == len(a)-1:
                return 'No'
        else:
            if stc==[]:
                return 'No'
            else:
                if stc[-1]==otk[zak.index(a[i])]:
                    stc.pop()
                else: return 'No'
    return 'Yes'
print(func(a))
```

```python
n=int(input())
N=[list(map(int, input().split())) for i in range(n)]
def povt(N):
    return N[1]
N.sort(key=povt)
last = N[0][1]
ans = 1
for l, r in N:
    if l > last:
        ans += 1
        last = r
print(ans)
```

Задача C

```python
def mera(A):
    c=0
    n=len(A)
    for i in range(n-1):
        for j in range(i+1, n):
            if A[i] > A[j]: c+=1
    return c
def sortirovka():
    n, m = map(int, input().split())
    M=[input() for i in range(m)]
    M.sort(key=mera)
    for i in M:
        print(i)
K=int(input())
for i in range(K):
    sortirovka()
    print()
```

Задача D

```python
n=int(input())
N=[list(map(int, input().split())) for i in range(n)]
def povt(N):
    return N[1], N[0]
def povtirz(N):
    return N[1], N[1]-N[0]
def poperirz(N):
    return N[0], N[1]-N[0]
K=sorted(N, key=poperirz)
L=sorted(N, key=povtirz)
M=sorted(N)
N.sort(key=povt)
def podsch(N):
    t=0
    c1=0
    flag = True
    while flag:
        flag = False
        c2=0
        for i in range(len(N)):
            if i-c2 > len(N)-1: break
            if t+N[i-c2][0] <= N[i-c2][1]:
                t+=N[i-c2][0]
                N.pop(i-c2)
                c2+=1
                flag = True
        c1+=c2
    return c1
print(max(podsch(N), podsch(M), podsch(L), podsch(K)))
```

```python
Data = list(map(int, input().split()))

def mediana(data):
   a=len(data)
   if a%2==0:
      print((data[a//2]+data[a//2-1]-1)//2+1, end=' ')
   else: print(data[a//2], end=' ')

data=[]
for i in Data:
   if not data:
      data.append(i)
   else:
      for j in range(len(data)):
         if data[j] < i:
            data.insert(j, i)
            break
      else: data.append(i)
   mediana(data)
```
