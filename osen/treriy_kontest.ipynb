Задача A

```python
def proverka(A):
    if None in A[:(len(A)-1)//2]:
        return print('NO')
    for i in range(0, (len(A)-1)//2):
        if A[2*i+1] and not A[i] > A[2*i+1]:
            return print('NO')
        if A[2*i+2] and not A[i] < A[2*i+2]:
            return print('NO')
    return print('YES')

def intnan(n):
    if n == 'None':
        return None
    else: return int(n)


A = list(map(intnan, input().split()))
proverka(A)
```

Задача B

```python
N=int(input())
Z=[list(input().split()) for i in range(N)]
mag=[]
art=[]
for i in Z:
    mag.append(i[0])
    art.append(i[1])

def vivod(n):
    if n in mag: print(art[mag.index(n)])
    else: print(mag[art.index(n)])

for i in range(int(input())):
    vivod(input())
```

Задача C

```python
t = int(input())
for a in range(t):
    n, q = map(int, input().split())
    N = list(map(int, input().split()))
    for b in range(q):
        i, j = map(int, input().split())
        i-=1
        c=0
        d=0
        M=N[i:j]
        for z in M:
            if z != d:
                a=M.count(z)
                if a > c:
                    c=a
                    d=z
        print(c)
```

Задача D

```python
def stuka():
    n = int(input())
    S=list(map(int, input().split()))
    p=0
    ans=-1
    while p <= n-1:
        s=0
        N=[]
        for z in range(p+1, n):
            if S[z] >= S[p]:
                N.append(S[z])
            else: break
                
        if N:
            s=S.index(max(N))-p
        if s != 0:
            p+=s
            if s > ans: ans = s
        else: p+=1
    print(ans)

for Z in range(int(input())):
    stuka()
```
