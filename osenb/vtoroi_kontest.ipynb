Задача A

```python
a=input()
def func(a):
    stc=[]
    otk=['(','[','{']
    zak=[')',']','}']
    for i in range(len(a)):
        if a[i] in otk:
            stc.append(a[i])
            if i == len(a)-1:
                return 'No'
        else:
            if stc==[]:
                return 'No'
            else:
                if stc[-1]==otk[zak.index(a[i])]:
                    stc.pop()
                else: return 'No'
    return 'Yes'
print(func(a))
```

```python
def mera(A):
    c=0
    n=len(A)
    for i in range(n-1):
        for j in range(i+1, n):
            if A[i] > A[j]: c+=1
    return c
def sortirovka():
    n, m = map(int, input().split())
    M=[input() for i in range(m)]
    M.sort(key=mera)
    for i in M:
        print(i)
K=int(input())
for i in range(K):
    sortirovka()
    print()
```

```python
n=int(input())
N=[list(map(int, input().split())) for i in range(n)]
def povt(N):
    return N[1], N[0]
def povtirz(N):
    return N[1], N[1]-N[0]
def poperirz(N):
    return N[0], N[1]-N[0]
K=sorted(N, key=poperirz)
L=sorted(N, key=povtirz)
M=sorted(N)
N.sort(key=povt)
def podsch(N):
    t=0
    c1=0
    flag = True
    while flag:
        flag = False
        c2=0
        for i in range(len(N)):
            if i-c2 > len(N)-1: break
            if t+N[i-c2][0] <= N[i-c2][1]:
                t+=N[i-c2][0]
                N.pop(i-c2)
                c2+=1
                flag = True
        c1+=c2
    return c1
print(max(podsch(N), podsch(M), podsch(L), podsch(K)))
```

```python
import heapq as hq
n=int(input())
N=[list(map(int, input().split())) for i in range(n)]
time=0
c=0
lt=[]
N.sort(key=lambda x: x[1])
for t, d in N:
    if time + t <= d:
        hq.heappush(lt, -t)
        time+=t
        c+=1
    else:
        if lt:
            if -lt[0] > t:
                max_t = -hq.heappushpop(lt, -t)
                time = time - max_t + t
print(c)
```

```python
def metod(A):
    return -len(A), A[0]
N = list(input().split())
M={}
for i in N:
    n = str(sorted(i))
    if not n in M.keys():
        M.update({n:[i]})
    else: M[n].append(i)
L=list(map(sorted, list(M.values())))
L.sort(key=metod)
for i in L:
    for j in i:
        print(j, end=' ')
```
