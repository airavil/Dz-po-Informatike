```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
a,b,c=int(input('множитель частоты x = ?')), int(input('множитель частоты y = ?')), int(input('смещение фазы y = п/?'))
t=np.arange(0, 100, 0.01)
x = np.sin(math.pi*t/a)
y = np.sin((math.pi*t+math.pi/c)/b)
plt.plot(x, y)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True, linestyle='dotted')
plt.title('Фигуры Лиссажу')
plt.show()
```

```python
plt.pie([0.01, 0.5, 0.3],autopct='%1.f%%' , labels = ['Да, сейчас этим займусь','Нет, я лучше инфу на пятницу заранее сделаю','Полежать под мягким одеяльцем с плюшевым мишкой'])
plt.title('Надо бы подготовить отчет по лабе')
plt.show()
```
