```python
A=input()
B=input()
visited=set()
a=[]
b=[]
ans=0
for i in range(len(A)):
    if A[i] != B[i]:
        a.append(A[i])
        b.append(B[i])
for i in range(len(a)):
    if i not in visited:
        visited.add(i)
        u=a[i]
        v=b[i]
        a[i]='X'
        b[i]='X'
        while u != v:
            j = a.index(v)
            if j not in visited:
                visited.add(j)
                x=b[j]
                a[j]='X'
                b[j]='X'
                ans+=1
                v=x
                
print(ans)
```

```python
A=input()
B=input()
ans=0
visited=set()
C=[]
for i in range(len(A)):
    C.append([A[i],B[i]])
while C:
    for i in C:
        if i[0]==i[1]:
            C.remove(i)
            break
    else:
        for i in C:
            if reversed(i) in C:
                ans+=1
                C.remove(i)
                C.remove(reversed(i))
                break
        else:
            for j in C:
                if j[0] == i[1]:
                    C.remove(i)
                    j[0]=i[0]
                    ans+=1
                    break
print(ans)
```

```python
import sys
sys.setrecursionlimit(100000)
n, m = map(int, input().split())
graph = [[] for i in range(n)]
visited = set()
count=0

def dfs(start, parent):
    global visited
    global cycle_
    if start in visited:
        return True
    visited.add(start)
    for u in graph[start]:
        if not cycle_ and u != parent:
            cycle_ = dfs(u, start)
    return cycle_


for i in range(m):
    x, y = map(int, input().split())
    x -= 1
    y -= 1
    graph[x].append(y)
    graph[y].append(x)

for v in range(n):
    if not v in visited:
        cycle_ = False
        cycle_ = dfs(v, -1)

        if not cycle_:
            count += 1

print(count)
```
