```python
import matplotlib.pyplot as plt
import networkx as nx
from collections import deque
```

Задача 1

```python
G = nx.Graph()
rebra = {
    (0, 1, 1),
    (0, 2, 1),
    (0, 3, 2),
    (0, 4, 3),
    (1, 4, 5),
    (2, 3, 2),
    (3, 4, 4)
}
for a, b, ves in rebra:
    G.add_edge(a, b, weight=ves)
plt.figure(figsize=(6,6))
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, node_color='pink')
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'weight'))
plt.show
```

Задача 2

```python
def dorogi(Graff, A, B):
    count=0
    visited = set()
    queue = deque([A])

    while queue:
        node = queue.popleft()
        if node not in visited:
            if node != B:
                visited.add(node)
                queue.extend(Graff[node])
            else:
                count+=1
    return count

G={
    "А":{"Б","В","Г","Д"},
    "Б":{"В","Е"},
    "В":{"Е","Ж"},
    "Г":{"В","Ж"},
    "Д":{"Г","Ж","З"},
    "Е":{"Ж","И"},
    "Ж":{"И"},
    "З":{"Ж","И"},
    "И":{}
}

print(dorogi(G, 'А', 'В')*dorogi(G, 'В', 'И') )
```
