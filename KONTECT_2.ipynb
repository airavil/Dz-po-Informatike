Задача A

```python
import heapq

n, m, k = map(int, input().split())
towns = list(map(int, input().split()))
graph = [[] for i in range(n + 1)]

for i in range(m):
    u, v, w = map(int, input().split())
    graph[u].append([v, w])
    graph[v].append([u, w])

INF = float('inf')
ans_dist = [0] * (n + 1)

def dijkstra(start):
    dist = [INF] * (n + 1)
    dist[start] = 0
    q = [(0, start)]
    while q:
        d, v = heapq.heappop(q)
        if d <= dist[v]:
            for to, w in graph[v]:
                x = d + w
                if x < dist[to]:
                    dist[to] = x
                    heapq.heappush(q, (x, to))

    return dist

for j in towns:
    dist = dijkstra(j)
    for i in range(1, n + 1):
        ans_dist[i] += dist[i]

city = min(range(1, n + 1), key=lambda i: (ans_dist[i], i))
print(city, ans_dist[city])
```

```python
from collections import deque
import heapq as hq


n, e, m = map(int, input().split())

INF = float('inf')
graph = {}
for i in range(n+1):
    graph.update({i : []})

for i in range(m):
    l = deque(map(int, input().split()))
    k = l.popleft()
    s1 = l.popleft()
    t1 = l.popleft()
    if s1 == 1:
        graph[0].append([s1, 0, t1])
    for j in range(k-1):
        s = l.popleft()
        t2 = l.popleft()
        graph[s1].append([s, t1, t2]) 
        s1= s
        t1= t2


pq = [(0, 0)]
min_v = [float('inf')]*(n+1)
min_v[0] = 0
while pq:
    curr_v, node = hq.heappop(pq)

    if curr_v > min_v[node]:
        continue
    
    for st, t1, t2 in graph[node]:
        if not t1 < curr_v:
            if  t2 < min_v[st]:
                min_v[st] = t2
                hq.heappush(pq, (t2, st))
if min_v[e] != INF:
    print(min_v[e])
else: print(-1)

```

```python
from collections import deque

class FordFulkerson:
    def __init__(self, rows, cols, blocked):
        self.X = rows
        self.Y = cols
        self.source = self.X + self.Y
        self.sink = self.X + self.Y + 1
        self.size = self.X + self.Y + 2

        self.graph = [[] for _ in range(self.size)]
        self.capacity = {}

        for i in range(self.X):
            self.graph[self.source].append(i)
            self.graph[i].append(self.source)
            self.capacity[(self.source, i)] = 1
            self.capacity[(i, self.source)] = 0

        for i in range(self.X):
            for j in range(self.Y):
                if (i,j) not in blocked:
                    u = i
                    v = self.X + j
                    self.graph[u].append(v)
                    self.graph[v].append(u)
                    self.capacity[(u, v)] = 1
                    self.capacity[(v, u)] = 0

        for j in range(self.Y):
            u = self.X + j
            v = self.sink
            self.graph[u].append(v)
            self.graph[v].append(u)
            self.capacity[(u, v)] = 1
            self.capacity[(v, u)] = 0

    def bfs(self, parent):
        visited = [False]*self.size
        queue = deque([self.source])
        visited[self.source] = True

        while queue:
            u = queue.popleft()
            for v in self.graph[u]:
                if not visited[v] and self.capacity.get((u, v), 0) > 0:
                    visited[v] = True
                    parent[v] = u
                    queue.append(v)
                    if v == self.sink:
                        return True
        return False

    def max_matching(self):
        parent = [-1]*self.size
        max_flow = 0

        while self.bfs(parent):
            path_flow = float('inf')
            v = self.sink
            while v != self.source:
                u = parent[v]
                path_flow = min(path_flow, self.capacity.get((u, v), 0))
                v = u

            v = self.sink
            while v != self.source:
                u = parent[v]
                self.capacity[(u, v)] -= path_flow
                self.capacity[(v, u)] = self.capacity.get((v, u), 0) + path_flow
                v = u

            max_flow += path_flow

        return int(max_flow)

tests = int(input())
for i in range(tests):
    rows, cols, cuts = map(int, input().split())
    if cuts > 0:
        cells = list(map(int, input().split()))
        ff = FordFulkerson(rows, cols, {(cells[i*2], cells[i*2+1]) for i in range(cuts)})
        print(ff.max_matching())
    else:
        input()
        print(min(rows, cols))
```

```python
class Edge:
    def __init__(self, v, flow, C, rev):
        self.v = v
        self.flow = flow
        self.C = C
        self.rev = rev


class Graph:
    def __init__(self, V):
        self.adj = [[] for _ in range(V)]
        self.V = V
        self.level = [0] * V

    def addEdge(self, u, v, C):
        a = Edge(v, 0, C, len(self.adj[v]))
        b = Edge(u, 0, 0, len(self.adj[u]))
        self.adj[u].append(a)
        self.adj[v].append(b)

    def BFS(self, s, t):
        self.level = [-1] * self.V
        self.level[s] = 0

        q = [s]
        while q:
            u = q.pop(0)
            for e in self.adj[u]:
                if self.level[e.v] < 0 and e.flow < e.C:
                    self.level[e.v] = self.level[u] + 1
                    q.append(e.v)

        return self.level[t] >= 0

    def sendFlow(self, u, flow, t, start):
        if u == t:
            return flow

        while start[u] < len(self.adj[u]):
            e = self.adj[u][start[u]]

            if self.level[e.v] == self.level[u] + 1 and e.flow < e.C:
                curr_flow = min(flow, e.C - e.flow)
                temp_flow = self.sendFlow(e.v, curr_flow, t, start)

                if temp_flow:
                    e.flow += temp_flow
                    self.adj[e.v][e.rev].flow -= temp_flow
                    return temp_flow

            start[u] += 1

        return 0

    def DinicMaxflow(self, s, t):
        if s == t:
            return 0

        for u in range(self.V):
            for e in self.adj[u]:
                e.flow = 0

        total = 0

        while self.BFS(s, t):
            start = [0] * self.V
            while True:
                flow = self.sendFlow(s, float('inf'), t, start)
                if not flow:
                    break
                total += flow

        return total

t = int(input())
for i in range(t):
    n = int(input())
    graph_ = [list(map(int, input())) for j in range(n)]
    answer = 10**10
    for s in range(n):
        for t_ in range(n):
            if s == t_:
                continue
            g = Graph(n)
            visited = set()
            for i in range(n):
                for j in range(n):
                    if i != j and (i, j) not in visited:
                        if graph_[i][j] > 0:
                            g.addEdge(i, j, graph_[i][j])
                            g.addEdge(j, i, graph_[i][j])
                        visited.add((i, j))
                        visited.add((j, i))
            flow = g.DinicMaxflow(s, t_)
            answer = min(answer, flow)

    print(answer)
```
