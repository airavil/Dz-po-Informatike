Задача A

```python
import heapq

n, m, k = map(int, input().split())
towns = list(map(int, input().split()))
graph = [[] for i in range(n + 1)]

for i in range(m):
    u, v, w = map(int, input().split())
    graph[u].append([v, w])
    graph[v].append([u, w])

INF = float('inf')
ans_dist = [0] * (n + 1)

def dijkstra(start):
    dist = [INF] * (n + 1)
    dist[start] = 0
    q = [(0, start)]
    while q:
        d, v = heapq.heappop(q)
        if d <= dist[v]:
            for to, w in graph[v]:
                x = d + w
                if x < dist[to]:
                    dist[to] = x
                    heapq.heappush(q, (x, to))

    return dist

for j in towns:
    dist = dijkstra(j)
    for i in range(1, n + 1):
        ans_dist[i] += dist[i]

city = min(range(1, n + 1), key=lambda i: (ans_dist[i], i))
print(city, ans_dist[city])
```

```python
from collections import deque
import heapq as hq


n, e, m = map(int, input().split())


graph = {}
for i in range(e+1):
    graph.update({i : set()})


for i in range(m):
    l = deque(map(int, input().split()))
    k = l.popleft()
    s1 = l.popleft()
    t1 = l.popleft()
    for j in range(k-1):
        s = l.popleft()
        t2 = l.popleft()
        if e <= s and s1 < e:
            graph[s1].add((s, t1, t2)) 
        s1= s
        t1= t2


pq = [(0, 0, 0)]
dist = [float('inf')]*(e+1)
dist[0] = 0
while pq:
    pribitie, otpravka , station = hq.heappop(pq)

    if pribitie > dist[station]:
        continue

    for neighbor, t1, t2 in graph[station]:
        if otpravka <= t1:
            vremy = t2
            if vremy < dist[neighbor]:
                dist[neighbor] = vremy
                hq.heappush(pq, (vremy+t2, vremy,neighbor))
            

print(dist[e])

```

```python
from collections import deque

class FordFulkerson:
    def __init__(self, rows, cols, blocked):
        self.X = rows
        self.Y = cols
        self.source = self.X + self.Y
        self.sink = self.X + self.Y + 1
        self.size = self.X + self.Y + 2

        self.residual_graph = [[0]*self.size for _ in range(self.size)]

        for i in range(self.X):
            self.residual_graph[self.source][i] = 1

        for i in range(self.X):
            for j in range(self.Y):
                if (i,j) not in blocked:
                    self.residual_graph[i][self.X + j] = 1

        for j in range(self.Y):
            self.residual_graph[self.X + j][self.sink] = 1

    def bfs(self, parent):
        visited = [False]*self.size
        queue = deque([self.source])
        visited[self.source] = True

        while queue:
            u = queue.popleft()
            for v in range(self.size):
                if not visited[v] and self.residual_graph[u][v] > 0:
                    visited[v] = True
                    parent[v] = u
                    queue.append(v)
                    if v == self.sink:
                        return True
        return False

    def max_matching(self):
        parent = [-1]*self.size
        max_flow = 0

        while self.bfs(parent):
            path_flow = float('inf')
            v = self.sink
            while v != self.source:
                u = parent[v]
                path_flow = min(path_flow, self.residual_graph[u][v])
                v = u

            v = self.sink
            while v != self.source:
                u = parent[v]
                self.residual_graph[u][v] -= path_flow
                self.residual_graph[v][u] += path_flow
                v = u

            max_flow += path_flow

        return int(max_flow)

tests = int(input())
for i in range(tests):
    rows, cols, cuts = map(int, input().split())
    blocked = set()
    if cuts > 0:
        cells = list(map(int, input().split()))
        for i in range(0, len(cells), 2):
            blocked.add((cells[i], cells[i+1]))
    else: input()

    ff = FordFulkerson(rows, cols, blocked)
    print(ff.max_matching())
```
